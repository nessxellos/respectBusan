```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from konlpy.tag import *
# csv       
data=pd.read_csv('insta_/insta_.csv')

tag_str=''
for i in data['']:
    #    
    if type(i) is float:
        continue
    tag_str+=i

# #      
taglist=tag_str.split('#')
# print(taglist)

# #       
str2=''
for i in taglist:
    str2+=i
    str2+=' '
# print(str2)

#    
okt = Okt()

#   
noun=okt.nouns(str2)
# print(noun)

tag_str=''
for i in noun:
    tag_str+=i
    tag_str+=' '
# print(tag_str)

#     
font_path = 'C:/Windows/Fonts/malgunsl.ttf'
wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path).generate(tag_str)
plt.figure(figsize = (10, 10))
plt.imshow(wc)
plt.axis("off")
wc.to_file("wordcloud/.jpg")
plt.show()
```
