```python
from wordcloud import WordCloud
import pandas as pd
from konlpy.tag import *
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
# csv       
data=pd.read_csv('/.csv')
tag_str=''
for i in data['']:
    #    
    if type(i) is float:
        continue
    tag_str += i
    
# #      
taglist=tag_str.split('#')

# #       
str2 = ''
for i in taglist:
    str2 += i
    str2 += ' '

#    
okt = Okt()
#   
nouns = okt.nouns(str2)

#       
tag_str = ''
for i in nouns:
    tag_str += i
    tag_str += ' '

#    . csv  .
#  
count = Counter(nouns)
#    
sorted_count = sorted(count.items(), key = lambda item: item[1], reverse=True)
#  
df = pd.DataFrame(sorted_count, columns = {'',''})
# csv  
df.to_csv('/.csv', index=False)

#     
# boogi     mask 
boogi_mask = np.array(Image.open('boogi_black.png'))
font_path = 'C:/Windows/Fonts/malgunsl.ttf'
wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path, mask = boogi_mask).generate(tag_str)
plt.figure(figsize = (10, 10))
plt.imshow(wc)
plt.axis("off")
wc.to_file("/.jpg")
plt.show()
```
