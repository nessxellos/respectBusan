```python
# 구글 관광지 리뷰 크롤링
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time
import os

path = 'C:/devtools/pythonwork/Lesson/chromedriver'
```

```python
# 구글 사이트 접근 환경 설정 완료
driver = webdriver.Chrome(path)
action = ActionChains(driver)
```

```python
# 검색할 관광지 입력
search_word = input('검색할 단어를 입력하세요 : ')
# review_n = input('가져올 리뷰의 숫자를 입력하세요 : ')

url = 'https://www.google.co.kr/maps/place/' + search_word + '/place'

# 검색한 관광지의 리뷰 화면 호출
driver.implicitly_wait(6)
driver.get(url)
#driver.maximize_window()ㅁ
#driver.find_element_by_xpath('/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[20]/div/div[2]/button').click()
```

```python
# 최초 호출
a = 1
# div[8] -> div[9]로 변경
reviews = driver.find_elements_by_xpath('/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[2]/div[8]/div['+ str(a) +']/div/div[3]/div[3]/div[2]')
scores = driver.find_elements_by_xpath('/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[2]/div[8]/div['+ str(a) +']/div/div[3]/div[3]/div[1]/span[2]')
#print(reviews.text)
#print(scores.get_attribute('aria-label'))
```

```python
# 리뷰와 별점을 담을 리스트 생성
reviews_list = []
scores_list = []
n = 1

# 리뷰를 리뷰리스트에 넣기
for i in reviews:
    #print(i.text)
    reviews_list.append(i.text)
    
# 별점을 별점 리스트에 넣기    
for i in scores:
    #print(i.text)
    scores_list.append(i.text)
    n += 1
    
# 반복
while True :
    try :
        if 130 > n : 
            driver.implicitly_wait(6)
            a += 3
            reviews = driver.find_elements_by_xpath('/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[2]/div[8]/div['+ str(a) +']/div/div[3]/div[3]/div[2]')
            scores = driver.find_elements_by_xpath('/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[2]/div[8]/div['+ str(a) +']/div/div[3]/div[3]/div[1]/span[2]')
            for i in reviews:
                print(i.text)
                reviews_list.append(i.text)  
                
            for i in scores:
                print(i.get_attribute('aria-label'))
                scores_list.append(i.get_attribute('aria-label'))
                n += 1
        
        else :
            if n >= 129 :
                break
            else : 
                driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    
    except :
        if n >= 129 :
            break    
```

```python
## 데이터프레임 생성
### 리뷰
review0 = pd.DataFrame()
review0[search_word] = reviews_list
### 별점
score0 = pd.DataFrame()
score0[search_word] = scores_list
## 두개의 데이터프레임 합친 후 저장
data = pd.merge(review0, score0, left_index = True, right_index = True)
data.columns = ['리뷰', '별점']
data.to_csv('reviews_송정해수욕장.csv', index=False)
data
```
