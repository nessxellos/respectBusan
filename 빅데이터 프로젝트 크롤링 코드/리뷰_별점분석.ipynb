```python
from wordcloud import WordCloud
import pandas as pd
from konlpy.tag import *
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
from konlpy.tag import Okt
from collections import Counter
import re
```

```python
# 리뷰가져오기
file = 'reviews_40계단(644)'
data=pd.read_csv('./리뷰/'+file+'.csv', names=['reviews','score'])

# 별표1, 2개 = Neg,  별표4, 5개 = Pos
df_Neg=data.loc[(data['score']==' 별표 1개 ')|(data['score']==' 별표 2개 '),['reviews']]
df_Pos=data.loc[(data['score']==' 별표 4개 ')|(data['score']==' 별표 5개 '),['reviews']]

#df_Neg.to_csv('Neg_'+file+'.csv', header = False, index = False, encoding='utf-8-sig')
#df_Pos.to_csv('Pos_'+file+'.csv', header = False, index = False, encoding='utf-8-sig')
```

```python
tag_Neg=''
tag_Pos=''

for i in df_Neg['reviews']:
    # 칼럼값이 없을 경우 처리
    if type(i) is float:
        continue
    tag_Neg+=i

for i in df_Pos['reviews']:
    # 칼럼값이 없을 경우 처리
    if type(i) is float:
        continue
    tag_Pos+=i    
    
#print(tag_Neg)
#print(tag_Pos)
```

```python
# 태그 리스트에서 형태소 분석하기

okt = Okt()

# 문자열에서 명사만 추출하기
noun_Neg=okt.nouns(tag_Neg)
noun_Pos=okt.nouns(tag_Pos)

# 명사만 추출한 결과를 다시 하나의 문자열로 만듦
tag_Neg2=''
for i in noun_Neg:
    tag_Neg2+=i
    tag_Neg2+=' '
    
tag_Pos2=''
for i in noun_Pos:
    tag_Pos2+=i
    tag_Pos2+=' '
```

```python
# 불필요 단어 제거 (원하는 단어 추가 가능)
tag_Pos3 = re.sub("원문|번역|제공|곳|부산|수","",tag_Pos2)
tag_Neg3 = re.sub("원문|번역|제공|곳|부산|수","",tag_Neg2)

# 단어를 공백을 기준으로 자름
tag_Pos_list = tag_Pos3.split()
tag_Neg_list = tag_Neg3.split()

# 단어 집계
count_Pos = Counter(tag_Pos_list)
count_Neg = Counter(tag_Neg_list)

# 값을 기준으로 내림차순 정렬
sorted_count_Pos = sorted(count_Pos.items(), key = lambda item: item[1], reverse=True)
sorted_count_Neg = sorted(count_Neg.items(), key = lambda item: item[1], reverse=True)

#print(sorted_count_Neg)
#print(sorted_count_Pos)
```

```python
# 데이터프레임으로 저장
df_pos = pd.DataFrame(sorted_count_Pos, columns = {'단어','빈도수'})
df_neg = pd.DataFrame(sorted_count_Neg, columns = {'단어','빈도수'})
# csv 파일 생성
df_pos.to_csv('./리뷰/positive/'+file+'.csv', header = True, index = False, encoding='utf-8-sig')
df_neg.to_csv('./리뷰/negative/'+file+'.csv', header = True, index = False, encoding='utf-8-sig')
```
